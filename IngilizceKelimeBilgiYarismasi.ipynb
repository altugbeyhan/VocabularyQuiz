```python
import random

def generate_question(used_questions):
    unused_questions = list(set(data.keys()) - set(used_questions))
    if not unused_questions:
        return None

    question_key = random.choice(unused_questions)
    used_questions.append(question_key)

    correct_answer = data[question_key]
    options = [correct_answer]
    while len(options) < 4:
        random_answer = random.choice(list(data.values()))
        if random_answer not in options:
            options.append(random_answer)

    random.shuffle(options)
    return question_key, options, options.index(correct_answer) + 1

def bilgi_yarismasi():
    print("Kelime Yarışması Oyununa Hoş Geldiniz!")

    score = 0
    num_questions = 5
    used_questions = []
    wrong_answers = []

    for i in range(num_questions):
        question_data = generate_question(used_questions)
        if question_data is None:
            print("Tüm kelimeleri doğru bildiniz! Oyun bitti.")
            break

        question, options, correct_option = question_data

        print(f"\nKelime {i + 1}/{num_questions}: {question}")
        for j, option in enumerate(options, start=1):
            print(f"{j}. {option}")
        try:
            user_choice = int(input("Cevap (1-4): "))

            if 1 <= user_choice <= 4:
                if user_choice == correct_option:
                    print("DOĞRU! :)")
                    score += 1
                else:
                    print(f"YANLIŞ! :( Doğru Cevap: {correct_option}. {options[correct_option - 1]}")
                    wrong_answers.append(question)

        except:
            print("Yanlış giriş.")
            wrong_answers.append(question)

    print(f"\nOyun bitti! {num_questions} kelimenin {score} tanesini doğru bildiniz.")

    if wrong_answers:
        print(f"\nTekrar Edilmesi Gereken {len(wrong_answers)} Kelime:")
        for wrong_answer in wrong_answers:
            print(wrong_answer,":",data[wrong_answer])

if __name__ == "__main__":
    bilgi_yarismasi()
```
